### Step-3 Converting Sentence into Embeddings
    * Download Glove 6B.50d.txt

```python
f = open('glove.6B.50d.txt',encoding ='utf-8')
```

```python
embeddings_index = {} # we created our own dictionary and copied the values from glove 6billion values to our dictionary
cnt = 0
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:],dtype='float')
    embeddings_index[word] = coefs
    
f.close()
    
```

```python
emb_dim = embeddings_index["eat"].shape[0] # embedding dimension which is of glove.txt is 50
print(emb_dim)
```

### step-4 Converting sentences into Vectors ** (Embedding Layer Output)

```python
def embedding_output(X):
    maxLen = 10
    embedding_out = np.zeros((X.shape[0],maxLen,emb_dim)) 
    
    for ix in range(X.shape[0]):
        
        X[ix] = X[ix].split() # it is for splitting the words from the sentences and then we'll look for the embbedding 
    
        for ij in range(len(X[ix])):
            # go to every word in the current (ix) sentence
            try:
                embedding_out[ix][ij] = embeddings_index[X[ix][ij].lower()]  # making in lower character
            except:
                embedding_out[ix][i]= np.zeros((50,))
                
    return embedding_out

embeddings_matrix_train = embedding_output(X_train)
embeddings_matrix_test = embedding_output(X_test)
            
        
    
```

```python
from keras.utils import to_categorical
```

```python
Y_train = to_categorical(train[1],num_classes=5)
Y_test = to_categorical(test[1],num_classes=5)

print(Y_train.shape)
print(Y_train[0])
```

#### step-5 Define the RNN/LSTM model Feeding the data to LSTM based Architecture

```python
from keras.models import Sequential
from keras.layers import *
```

```python
model = Sequential()
model.add(LSTM(64,input_shape=(10,50),return_sequences=True)) # stacked LSTM 1st layer which returns sequencces
model.add(Dropout(0.5))
model.add(LSTM(64,return_sequences=False)) # 2nd lstm layer which donot return sequences
model.add(Dropout(0.5))
model.add(Dense(5))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint("best_model.h5",monitor="val_loss",verbose=True,save_best_only=True)
early_stop = EarlyStopping(monitor='val_acc',patience=10) # patience=10 if after 10 epochs val_accuracy not improves we'll stop

hist = model.fit(embeddings_matrix_train,Y_train,epochs=100,batch_size=64,shuffle=True,validation_split=0.2)
```

```python
pred = model.predict_classes(embeddings_matrix_test)
```
